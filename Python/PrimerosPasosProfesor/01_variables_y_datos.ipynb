## Usando variables

Definición de variable:
 - Caja donde se pueden almacenar valores (1a definición)
 - En Python, una variable es una etiqueta a la que podemos asignar valores y hace referencia a un valor determinado.

Reglas para nombrar variables:
 - Sólo pueden contener letras, números y guiones bajos.
 - No están permitidos los espacios, es útil seguir el método CamelCase.
 - Evitar el uso de palabras clave y nombres de función propios del lenguaje.
 - Usar nombres cortos pero descriptivos
 - Python es Case Sensitive, distingue minúsculas de mayúsculas.

```python
nombre = 'Manu'
print(nombre)

mensaje = 'Bienvenido al curso de Python en CIFO Barcelona La Violeta'
print(mensaje)
```

## Tipos de datos

En Python tenemos los siguientes tipos de datos simples:
- Cadenas de carácteres (strings)
- Números enteros
- Números decimales o de coma flotante
- Lógicos o booleanos (True o False)

Si queremos saber el tipo de dato que contiene cualquier variable, podemos hacer uso de la función *type(variable)*

```python
# Ejemplo de uso
nombre = 'Manu'
print(type(nombre))

numero = 12
print(type(numero))
```

## Comentarios en nuestro código

Los comentarios en Python se pueden realizar de dos formas:
- Usamos '#' para comentarios de una línea
- Usamos ''' ''' para bloques de comentarios

```python
# Comentario de una línea
'''
    Bloque
    de
    comentario
    de más de una línea
'''
```
