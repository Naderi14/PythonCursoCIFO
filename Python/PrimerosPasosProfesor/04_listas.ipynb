# Colecciones de datos

## Listas

Las listas son colecciones de elementos dispuestos en un orden en particular. En Python se usan corchetes [] para indicar que esa colección es una lista

```python
# Ejemplos de listas
numeros = [1, 2, 3, 4]
datos = [4, 'Una cadena', -15, 3.14, True]
print(numeros)
print(datos)
```

```python
# Ejemplo. Lista con distintos tipos de bicicletas
bicycles = ['trek', 'cannondale', 'redline', 'specialized']
print(bicycles)
```

Acceso a los elementos de una lista, mediante su índice o posición. La primera posición siempre tiene el índice '0'

```python
print(bicycles[0])
```

Para devolver el último elemento de la lista, podemos usar el índice negativo

```python
print(bicycles[-1].capitalize())
```

```python
# Más combinaciones
message = f'Mi primera bicicleta fue un model {bicycles[0].capitalize()}'
print(message)
```

### Modificando listas

Para modificar elementos de una lista (las listas son mutables), usamos el nombre de la lista seguido del índice del elemento a modificar y proporcionamos el nuevo valor

```python
# Ejemplo
motorcycles = ['Indian', 'Honda', 'Yamaha', 'Suzuki']
print(motorcycles)
```

```python
motorcycles[3] = 'Harley Davidson'
print(motorcycles)
```

Para añadir elementos al final de la lista (adjuntar), usamos el método append()

```python
motorcycles.append('Ducati')
print(motorcycles)
```

Para añadir elementos en cualquier posición (insertar), usamos el método .insert(), especificando su índice y valor

```python
motorcycles.insert(1, 'Triumph')
print(motorcycles)
```

### Eliminando elementos de una lista

Caso 1. Conocemos su índice y usando la sentencia *del*

```python
del motorcycles[0]
print(motorcycles)
```

Caso 2. Con el método pop(), eliminamos el último elemento de una lista y además podemos hacer uso de ese valor

```python
print(motorcycles)
```

```python
motorcyclePopped = motorcycles.pop()
print(motorcycles)
print(motorcyclePopped)
```

```python
motorcyclePopped2 = motorcycles.pop(2)
print(motorcycles)
print(motorcyclePopped2)
```

Caso 3. Conociendo el valor de la posición, podemos usar el método .remove(). Al igual que con el método .pop(), podemos hacer uso de su valor

```python
tooExpensive = 'Harley Davidson'
motorcycles.remove(tooExpensive)
print(motorcycles)
print(f'Una {tooExpensive} es demasiado cara para mí')
```

### Ordenando listas

Ordenar una lista de forma permanente con el método sort().
Hay que tener en cuenta las mayúsculas

```python
coches1 = ['seat', 'bmw', 'subaru', 'Renault']
coches1.sort()
print(coches1)
```

```python
coches2 = ['seat', 'bmw', 'subaru', 'renault']
coches2.sort()
print(coches2)
```

Orden alfabético inverso añadiendo el argumento reverse=True a sort()

```python
coches2.sort(reverse=True)
print(coches2)
```

Ordenar una lista temporalmente con la función sorted()

```python
motocicletas = ['bmw', 'royal enfield', 'triumph', 'yamaha', 'honda']
print(sorted(motocicletas))
print(motocicletas)
```

```python
# Orden temporal inverso
print(sorted(motocicletas, reverse=True))
```

Otros métodos

```python
# .reverse() invierte el orden de la lista
print(motocicletas)
```

```python
motocicletas.reverse()
print(motocicletas)
```

```python
# len(lista), muestra la longitud o cantidad de elementos que tiene
print(len(motocicletas))
```

### Listas dentro de listas (listas anidadas)

Podemos manipular este tipo de estructuras usando múltiples índices, como si nos refiriéramos a las filas y columnas de una tabla

```python
# Ejemplo
a = [1,2,3]
b = [4,5,6]
c = [7,8,9]
r = [a,b,c]
```

```python
print(r)
```

```python
# Accedemos a la primer sublista
r[0]
```

```python
# Accedemos a la última sublista
r[-1]
```

```python
# Accedemos al primer ítem de la primera sublista
r[0][0] 
```

```python
# Accedemos al último ítem de la última lista usando índices negativos
r[-1][-1]
```

### Trabajando con listas

#### Bucle For para recorrer listas

```python
magos = ['gandalf', 'merlín el encantador', 'raistlin', 'fistandantilus']
for mago in magos:
    print(mago)

print(mago)
```

```python
print(magos)
```

```python
# Salida más elaborada
for mago in magos:
    print(f'{mago.capitalize()}, qué gran truco de magia')
```

#### Listas numéricas

Uso de la función range(inicio, final-1).
Esta función genera una serie de números entre dos valores, el segundo siempre será valor-1

```python
# Ejemplo
for value in range(1,11):
    print(value, end=",")
```

Generamos una lista de números con la función range() y la función list()

```python
valores = list(range(1,11))
print(valores)
```

Si a la función range() le añadimos un tercer argumento, éste es el paso de valor

```python
# Imprimimos los números pares entre 0 y 10
pares = list(range(2,10,2))
print(pares)
```

```python
# Otro ejemplo. Calculamos los primeros diez cuadrados
cuadrados = []
for valores in range(1,11):
    cuadrado = valores ** 2
    cuadrados.append(cuadrado)

print(cuadrados)
```

Funciones estadísticas: min(), max(), sum()

```python
print(min(cuadrados))
print(max(cuadrados))
print(sum(cuadrados))
```

Listas por compresión (OJO!!!!)

```python
# nombreLista = [expresión bucleFor]
nuevosCuadrados = [valor**2 for valor in range(1,11)]
print(nuevosCuadrados)
```

#### Slicing de listas

Para dividir una lista es preciso especificar los índices de primer y último elemento que queremos trabajar. Se detiene en uno antes del último, igual que con la función range()

```python
# Ejemplo
jugadores = ['carlos', 'martina', 'miguel', 'florencio', 'elizabeth']
```

```python
# estraemos los tres primeros elementos, del 0 al 2
print(jugadores[0:3])
```

Si omitimos el primer índice, siempre va a empezar en la posición 0

```python
print(jugadores[:4])
```

Si omitimos el segundo índice, acabará en el final de la lista

```python
print(jugadores[2:])
```

También podemos usar índices negativos

```python
# extraer los tres últimos elementos
print(jugadores[-3:])
```

Si pasamos un tercer valor, indicamos el salto entre elementos

```python
print(jugadores[0:4:2])
```

Pasando un bucle por un slice

```python
print('Aquí tenemos los tres primeros jugadores de mi equipo')
for jugador in jugadores[:3]:
    print(jugador.title(), end=",")
    
```

#### Obteniendo datos de una lista

Podemos obtener el valor y el índice de una lista con la función *enumarate()*

```python
alumnos = ['Daniel', 'Domingo', 'Fran', 'Elizabeth', 'María']

for i, alumno in enumerate(alumnos):
    print(f'Alumno número {i+1}: {alumno}')

for i in enumerate(alumnos):
    print(i)
```
