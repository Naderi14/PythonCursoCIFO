## Listas

Las listas son colecciones de elementos dispuestos en un orden en particular. En Python se usan corchetes [] para indicar que esa coleccin es una lista

```python
# Ejemplos de listas
numeros = [1, 2, 3, 4]
datos = [4, 'Una cadena', -15, 3.14, True]
print(numeros)
print(datos)
```

```python
# Ejemplo. Lista con distintos tipos de bicicletas
bicycles = ['trek', 'cannondale', 'redline', 'specialized']
print(bicycles)
```

Acceso a los elementos de una lista, mediante su ndice o posicin. La primera posicin siempre tiene el ndice '0'

```python
print(bicycles[0])
```

Para devolver el ltimo elemento de la lista, podemos usar el ndice negativo

```python
print(bicycles[-1].capitalize())
```

```python
# Ms combinaciones
message = f'Mi primera bicicleta fue un model {bicycles[0].capitalize()}'
print(message)
```

### Modificando listas

Para modificar elementos de una lista (las listas son mutables), usamos el nombre de la lista seguido del ndice del elemento a modificar y proporcionamos el nuevo valor

```python
# Ejemplo
motorcycles = ['Indian', 'Honda', 'Yamaha', 'Suzuki']
print(motorcycles)
```

```python
motorcycles[3] = 'Harley Davidson'
print(motorcycles)
```

Para aadir elementos al final de la lista (adjuntar), usamos el mtodo append()

```python
motorcycles.append('Ducati')
print(motorcycles)
```

Para aadir elementos en cualquier posicin (insertar), usamos el mtodo .insert(), especificando su ndice y valor

```python
motorcycles.insert(1, 'Triumph')
print(motorcycles)
```

### Eliminando elementos de una lista

Caso 1. Conocemos su ndice y usando la sentencia *del*

```python
del motorcycles[0]
print(motorcycles)
```

Caso 2. Con el mtodo pop(), eliminamos el ltimo elemento de una lista y adems podemos hacer uso de ese valor

```python
print(motorcycles)
```

```python
motorcyclePopped = motorcycles.pop()
print(motorcycles)
print(motorcyclePopped)
```

```python
motorcyclePopped2 = motorcycles.pop(2)
print(motorcycles)
print(motorcyclePopped2)
```

Caso 3. Conociendo el valor de la posicin, podemos usar el mtodo .remove(). Al igual que con el mtodo .pop(), podemos hacer uso de su valor

```python
tooExpensive = 'Harley Davidson'
motorcycles.remove(tooExpensive)
print(motorcycles)
print(f'Una {tooExpensive} es demasiado cara para m')
```

### Ordenando listas

Ordenar una lista de forma permanente con el mtodo sort().
Hay que tener en cuenta las maysculas

```python
coches1 = ['seat', 'bmw', 'subaru', 'Renault']
coches1.sort()
print(coches1)
```

```python
coches2 = ['seat', 'bmw', 'subaru', 'renault']
coches2.sort()
print(coches2)
```

Orden alfabtico inverso aadiendo el argumento reverse=True a sort()

```python
coches2.sort(reverse=True)
print(coches2)
```

Ordenar una lista temporalmente con la funcin sorted()

```python
motocicletas = ['bmw', 'royal enfield', 'triumph', 'yamaha', 'honda']
print(sorted(motocicletas))
print(motocicletas)
```

```python
# Orden temporal inverso
print(sorted(motocicletas, reverse=True))
```

Otros mtodos

```python
# .reverse() invierte el orden de la lista
print(motocicletas)
```

```python
motocicletas.reverse()
print(motocicletas)
```

```python
# len(lista), muestra la longitud o cantidad de elementos que tiene
print(len(motocicletas))
```

### Listas dentro de listas (listas anidadas)

Podemos manipular este tipo de estructuras usando mltiples ndices, como si nos refiriramos a las filas y columnas de una tabla

```python
# Ejemplo
a = [1,2,3]
b = [4,5,6]
c = [7,8,9]
r = [a,b,c]
```

```python
print(r)
```

```python
# Accedemos a la primer sublista
r[0]
```

```python
# Accedemos a la ltima sublista
r[-1]
```

```python
# Accedemos al primer tem de la primera sublista
r[0][0] 
```

```python
# Accedemos al ltimo tem de la ltima lista usando ndices negativos
r[-1][-1]
```

#### Listas numricas

Uso de la funcin range(inicio, final-1).
Esta funcin genera una serie de nmeros entre dos valores, el segundo siempre ser valor-1

```python
# Ejemplo
for value in range(1,11):
    print(value, end=",")
```

Generamos una lista de nmeros con la funcin range() y la funcin list()

```python
valores = list(range(1,11))
print(valores)
```

Si a la funcin range() le aadimos un tercer argumento, ste es el paso de valor

```python
# Imprimimos los nmeros pares entre 0 y 10
pares = list(range(2,10,2))
print(pares)
```

```python
# Otro ejemplo. Calculamos los primeros diez cuadrados
cuadrados = []
for valores in range(1,11):
    cuadrado = valores ** 2
    cuadrados.append(cuadrado)

print(cuadrados)
```

Funciones estadsticas: min(), max(), sum()

```python
print(min(cuadrados))
print(max(cuadrados))
print(sum(cuadrados))
```

Listas por compresin (OJO!!!!)

```python
# nombreLista = [expresin bucleFor]
nuevosCuadrados = [valor**2 for valor in range(1,11)]
print(nuevosCuadrados)
```

#### Slicing de listas

Para dividir una lista es preciso especificar los ndices de primer y ltimo elemento que queremos trabajar. Se detiene en uno antes del ltimo, igual que con la funcin range()

```python
# Ejemplo
jugadores = ['carlos', 'martina', 'miguel', 'florencio', 'elizabeth']
```

```python
# estraemos los tres primeros elementos, del 0 al 2
print(jugadores[0:3])
```

Si omitimos el primer ndice, siempre va a empezar en la posicin 0

```python
print(jugadores[:4])
```

Si omitimos el segundo ndice, acabar en el final de la lista

```python
print(jugadores[2:])
```

Tambin podemos usar ndices negativos

```python
# extraer los tres ltimos elementos
print(jugadores[-3:])
```

Si pasamos un tercer valor, indicamos el salto entre elementos

```python
print(jugadores[0:4:2])
```

Pasando un bucle por un slice

```python
print('Aqu tenemos los tres primeros jugadores de mi equipo')
for jugador in jugadores[:3]:
    print(jugador.title(), end=",")
    
```

#### Obteniendo datos de una lista

Podemos obtener el valor y el ndice de una lista con la funcin *enumarate()*

```python
alumnos = ['Daniel', 'Domingo', 'Fran', 'Elizabeth', 'Mara']

for i, alumno in enumerate(alumnos):
    print(f'Alumno nmero {i+1}: {alumno}')

for i in enumerate(alumnos):
    print(i)
```
